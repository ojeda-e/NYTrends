Function requesting NYT archive via API 

```python
def my_archive(year, month, key):
    "Calls the URL and returns a json of the search results"
        
    url = (URL_ARCHIVE.format(year, month, key))
    print("Checking url ... ", url)
    r = requests.get(url)

    return r.json()['response']
```

### Iterate over months and years
Now let's create a function to iterate over months and years:

```python
from collections import defaultdict
word_freq = defaultdict(int)
word_freq["a"]+=1
def get_wordlist(top, month, year, my_key):
    """ Return a dictionary {"word": frequency} """

    print(month, year)
    list_words_inheader = []
    
    data = my_archive(year, month, my_key)
    
    for element in range(len(data['docs'])):
        if 'main' in data['docs'][element]['headline']:
        #print(jdata['docs'][element]['headline']['main'].split())
            list_words_inheader.extend(data['docs'][element]['headline']['main'].split() )
        else:
            print("No main for header, index", element)

    
    # Delete multiple occurrance
    new_list = list(set(list_words_inheader))  
    
        
    #Remove articles, personal pronouns, prepositions
    drop = ["the", "a", "A", "As", "as", "at", "At", "Against", "About", "By", "With", "It", "Is", "to", "To", "TO", "in", "In", "MARKETS:",
        "BRIEFING", "IN", "Into", "for", "For", "THE", "The", "An", "an", "and", "And", "Are", "After", "Of", "OF","of", "Off", "May",
        "On", "on", "or", 'OR', 'Out', "One", "Over", "by", "her", "his", "His", "From", "That", "This", "No", "but", "GUIDE",
        "But", "Not", "Paid", "Notice:", "Deaths", "New", "Corrections", "Memorials", "Back", "October", "First", "York", "You", "Your", "Times;",
        "Their", "It's", "Its", "How", "Two",  "Up", "Who", "What", "When", "Where", "We", "They", "vs.",
        "Journal.", "Books", "Sports", "Big", "Say", "Do", "Be", "Sports", "Will", "I"]

    
    #Remove number in range [0,100]
    drop.extend(str(i) for i in range(101))
    drop.extend(letter+'.' for letter in ascii_uppercase)
    drop.extend(month_name[i] for i in range(1,13))
    drop.extend(string.printable[i] for i in range(62,94))
    
    #print("Dropping elements:", drop)
    
    for word in drop:
        new_list2 = list(filter(lambda i: i != word, new_list))

#     for word in drop:
#         word_freq.pop(word, None)
   
    #print(new_list2)
    #create dictionary dict-like
    
    for i in new_list2:
        word_freq = {key : 0 for key in new_list2}

    #count frequency of each word
    for key in word_freq.keys():
        word_freq[key] = word_freq[key]+1
        
    #max(word_freq_.items(), key=operator.itemgetter(1))[0]
    
    fwords = {key: word_freq[key] for key in nlargest(top, word_freq, key = word_freq.get)}
    


    return fwords
```

And run it

```python
years = range(2000,2010)

wf_by_year = {key:[] for key in years}

for year in years:
    
    wf_by_year[year] = {month_name[key]: [] for key in range(1,13)}
    print('**Year:',year)#,wf_by_year[year]) 
    
    for index_month in range(1,13):
        print("Extracting data from", month_name[index_month], year)
        wf_by_year[year][month_name[index_month]] = get_wordlist(40, index_month, year, my_key)
```
